```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset('titanic')
```

# 1. Logistic Regression

```python
df.head()
```

```python
df.columns
```

```python
df.info() #deck have many null values 
```

```python
df_drop=df.drop([ 'class', 'who', 'adult_male', 'deck', 'embark_town','alive'],axis=1)
```

```python
df_drop
```

```python
df_drop.info()
```

```python
df_drop['age'].fillna(df['age'].mean(),inplace=True)
```

```python
df_drop.dropna(subset=['embarked'],inplace=True)
```

```python
df_drop.info()
```

### Label encode

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df_drop.head()
```

```python
df_drop['sex'] = le.fit_transform(df_drop['sex'])
df_drop['embarked'] = le.fit_transform(df_drop['embarked']) #S=2, C=0, Q=1
```

```python
df_drop.head()
```

```python
final_df=df_drop
```

```python
final_df['age']=final_df['age'].astype(int)
```

```python
final_df
```

### Train test Split 

```python
from sklearn.model_selection import train_test_split
```

```python
X=final_df.drop(['survived'],axis=1)
y=final_df['survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model =LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

#### Logistric Regression >> Classification Problem (Use Confusion Matrix for Evaluation)

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

# 2. KNN Classifier
